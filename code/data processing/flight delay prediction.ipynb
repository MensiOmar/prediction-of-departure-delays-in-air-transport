```python
import pytz
from datetime import datetime , timedelta
```

```python
tz_df = airports_data.loc[:,['IATA_CODE','tz']]
tz_df.rename(columns={'IATA_CODE':'ORIGIN_AIRPORT'},inplace=True)
data = flights_data
merged_data = pd.merge(tz_df,data, on='ORIGIN_AIRPORT')
merged_data.head()
```

```python
def add_time(row):
    if(row.ORIGIN_AIRPORT!='ISN'):
        hour = row['SCHEDULED_DEPARTURE'] // 100
        minute = row['SCHEDULED_DEPARTURE'] % 100
        time_zone = pytz.timezone(row['tz'])
        t = datetime(row['YEAR'], row['MONTH'], row['DAY'], hour, minute,0,tzinfo=time_zone)
        t_utc = t.astimezone(pytz.utc)
        
    else:
        hour = row['SCHEDULED_DEPARTURE'] // 100
        minute = row['SCHEDULED_DEPARTURE'] % 100
        if (row['MONTH'] in [11,12,1,2] and (row['MONTH']==3 and row['DAY']<8)):
            time_zone = pytz.timezone(row['tz'][1])
        else:
            time_zone = pytz.timezone(row['tz'][0])
        t = datetime(row['YEAR'], row['MONTH'], row['DAY'], hour, minute,0,tzinfo=time_zone)
        t_utc = t.astimezone(pytz.utc)
        
    return t_utc

merged_data["time"] = merged_data.apply(add_time, axis=1)
merged_data.loc[merged_data.ORIGIN_AIRPORT.isin(['ISN'])]
```

```python
def tolerance_delay(row):
    row.DEPARTURE_DELAY = row.DEPARTURE_DELAY - 15
    if row.DEPARTURE_DELAY <= 0:
        row.DEPARTURE_DELAY=0
    return row

merged_data=merged_data.apply(tolerance_delay, axis='columns')
```

```python
merged_data.dropna(subset=['DEPARTURE_DELAY'], inplace=True)
# Convert 6 hours to minutes
max_delay_minutes = 6 * 60

# Drop rows where 'DEPARTURE_DELAY' is greater than 6 hours (360 minutes)
merged_data = merged_data[merged_data['DEPARTURE_DELAY'] <= max_delay_minutes]
```

```python

dep_delay_max = merged_data["DEPARTURE_DELAY"].max()
dep_delay_min = merged_data["DEPARTURE_DELAY"].min()
# min = 0.0
# max = 360.0
merged_data['DEPARTURE_DELAY'] = merged_data['DEPARTURE_DELAY'].apply(lambda x: (x - 0.0) / (360.0 - 0.0))
merged_data["DEPARTURE_DELAY"].max()
```

```python
merged_data['year'] = merged_data['time'].dt.year
merged_data['month'] = merged_data['time'].dt.month
merged_data['day'] = merged_data['time'].dt.day
merged_data['hour'] = merged_data['time'].dt.hour
day_of_month={1:31,
              2:28,
              3:31,
              4:30,
              5:31,
              6:30,
             7:31,
             8:31,
             9:30,
             10:31,
             11:30,
             12:31}

yearly_groups = merged_data.groupby('year')
airports = airports_data.IATA_CODE.unique()
avg_delay_df = pd.DataFrame(columns=airports)

monthly_groups = yearly_groups.get_group(2015).sort_values(by='time').groupby('month')
for j in range(1,13):
    daily_groups = monthly_groups.get_group(j).sort_values(by='time')
    available_days = daily_groups['day'].unique()
    daily_groups = monthly_groups.get_group(j).sort_values(by='time').groupby('day')
    for k in range(1,day_of_month[j]+1):
        if (k not in available_days):
            continue
        else:
            hourly_groups = daily_groups.get_group(k).sort_values(by='time')
            available_hours = hourly_groups['hour'].unique()
            hourly_groups = daily_groups.get_group(k).sort_values(by='time').groupby('hour')

            for l in range(0,24):
                if (l not in available_hours):
                    continue
                else:
                    airport_groups=hourly_groups.get_group(l).sort_values(by='time')
                    available_airports = airport_groups['ORIGIN_AIRPORT'].unique()
                    airport_groups=hourly_groups.get_group(l).sort_values(by='time').groupby("ORIGIN_AIRPORT")

                    t = datetime(2015, j, k, l, 0,0,tzinfo=pytz.utc)
                    dict={'time':t}
                    for n in airports: 
                        if (n in available_airports):
                            dict[n]= airport_groups.get_group(n)['DEPARTURE_DELAY'].mean()
                        else:
                            dict[n]=0
                    new_row = pd.DataFrame(dict,index=[0])
                    avg_delay_df =  pd.concat([avg_delay_df,new_row], ignore_index=True)
                        

avg_delay_df
```
